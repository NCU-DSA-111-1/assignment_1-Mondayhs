```python
import csv
from ctypes import sizeof
import numpy as np
%pylab inline
from matplotlib import pyplot as plt
```

```python
def plt_loss(path,nbit):
    with open(path, newline='') as csvfile:        #newline='' 參數，這是為了讓資料中包含的換行字元可以正確被解析
  # 讀取 CSV 檔案內容
        reader = csv.reader(csvfile)
        data=list(reader)#,delimiter=",")
    data_range=size(data)/2
    epoch=np.array(np.int8([data[i][0] for i in range(1,int(data_range))]))
    error=np.array(np.float64([data[i][1] for i in range(1,int(data_range))]))
    plt.plot(error,'bo-')
    plt.grid()
    plt.title(f'{nbit} bits XOR')
    plt.xlabel('No. of Epochs/10')
    plt.ylabel('Error')
```

```python
plt.figure(figsize=(13,4))
plt.subplot(1,2,1)
path='xor_2bit.csv'
plt_loss(path,2)

plt.subplot(1,2,2)
path='xor_3bit.csv'
plt_loss(path,3)
```
